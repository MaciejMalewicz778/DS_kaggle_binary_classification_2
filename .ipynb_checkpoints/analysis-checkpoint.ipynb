### Some categorical are binary so lets convert them into 0 - 1 values (then they will become 'numerical')

```python
data_train['surgery'] = data_train['surgery'].apply(lambda x: 1 if x == 'yes' else 0)
data_train['is_young'] = data_train['age'].apply(lambda x: 1 if x == 'young' else 0)
data_train['surgical_lesion'] = data_train['surgical_lesion'].apply(lambda x: 1 if x == 'yes' else 0)
data_train['cp_data'] = data_train['cp_data'].apply(lambda x: 1 if x == 'yes' else 0)

new_numerical.extend(['surgery', 'is_young', 'surgical_lesion', 'cp_data'])

data_train.drop(columns=['age'], inplace=True)
```

### Either lesion_2 and lesion_3 should be removed from the dataframe (because most of the values are the same) or make feature descriptive (is_lesion_2, is_lesion_3).
### The distribution of total_protein and lesion_2 is also suspicious but at now, i will leave it as it is.

```python
data_train['is_lesion_2'] = data_train['lesion_2'].apply(lambda x: 0 if x == 0 else 1)
data_train['is_lesion_3'] = data_train['lesion_3'].apply(lambda x: 0 if x == 0 else 1)

new_numerical.extend(['is_lesion_2', 'is_lesion_3'])

data_train.drop(columns=['lesion_2', 'lesion_3'], inplace=True)
```

### Model building

```python
from sklearn.model_selection import train_test_split

# remove dependent variable and the one that i do not want to use during predictions
X = data_train.copy().drop(columns=['outcome', 'hospital_number', 'id'])
y = data_train.copy()['outcome']
X_d = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X_d, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Scale numerical
# remember that i created is_lesion_2 and is_lesion_3 which are considered as numerical. We do not scale them.
num_cols = X.select_dtypes(include=[np.number]).drop(columns=new_numerical).columns

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[num_cols] = scaler.fit_transform(X_train_scaled[num_cols])
X_test_scaled[num_cols] = scaler.transform(X_test_scaled[num_cols])
```

### In this notebook i will try to tune parameters

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier

# Define the models
models = {
    'RandomForestClassifier': RandomForestClassifier(random_state=17),
    'LogisticRegression': LogisticRegression(multi_class='multinomial', max_iter=1000, solver='lbfgs'),
    'GradientBoostingClassifier': GradientBoostingClassifier(random_state=17)
}

best_scores = {
    'RandomForestClassifier': {},
    'LogisticRegression': {},
    'GradientBoostingClassifier': {}
}

# Define the hyperparameter grids for each model
param_grids = {
    'RandomForestClassifier': {
        'n_estimators': [100, 200, 500],
        'max_depth': [None, 10, 30],
        'min_samples_split': [2, 5, 10],
        'criterion': ['log_loss', 'gini']
    },
    'LogisticRegression': {},
    'GradientBoostingClassifier': {
    'n_estimators': [100, 200, 500],
    'learning_rate': [0.01, 0.1, 0.3],
    'max_depth': [3, 6, 10],
    }
}
```

```python
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score

cv = KFold(n_splits=4, shuffle=True, random_state=17)

grids = {}
for model_name, model in models.items():
    grids[model_name] = GridSearchCV(estimator=model, param_grid=param_grids[model_name], scoring='f1_macro', cv=4, n_jobs=-1, verbose=2)
    
    grids[model_name].fit(X_train_scaled, y_train)
    best_scores[model_name]['best_params'] = grids[model_name].best_params_
    best_scores[model_name]['best_score'] = grids[model_name].best_score_
    print(f'model {model_name}')
    print(f'Best parameters for {model_name}: {best_scores[model_name]}')
```

### As we can see RandomForestClassifier produces very similar score to GradientBoostingClassifier
### But let's look at the classifiers closer

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier

best_models = {
    'RandomForestClassifier': RandomForestClassifier(random_state=17, max_depth=None, min_samples_split=2, n_estimators=200),
    'LogisticRegression': LogisticRegression(multi_class='multinomial', max_iter=1000, solver='lbfgs'),
    'GradientBoostingClassifier': GradientBoostingClassifier(random_state=17, learning_rate=0.3,max_depth=6, n_estimators=100)
}

for model_name, model in models.items():
    model.fit(X_train_scaled, y_train)
```

```python
preds = {}
for model_name, model in models.items():
    preds[model_name] = model.predict(X_test_scaled)
```

```python
# confusion matrices with the help of chatGPT
from sklearn.metrics import confusion_matrix

labels = ['died', 'lived', 'euthanized']

for model_name, model in models.items():
    print(f'Confusion matrix for {model_name}:')
    cm = confusion_matrix(y_test, preds[model_name], labels=labels)
    cm_df = pd.DataFrame(cm, index=['Actual died', 'Actual lived', 'Actual euthanized'], columns=['predicted died', 'predicted lived', 'predicted euthanized'])
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
```
